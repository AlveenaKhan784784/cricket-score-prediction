```python
df['city'] = df['city'].mask(df['city'].isnull(),df['venue'].str.split().str.get(0))
```

```python
df.isnull().sum()
```

```python
df.drop('venue', inplace=True, axis=1)
```

```python
eligible_cities = df['city'].value_counts()
```

```python
eligible_cities = eligible_cities[eligible_cities >= 600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df.head(15)
```

```python
df['over'] = df['ball'].astype(int)
```

```python
df.head(15)
```

```python
df['ball_no'] = df['ball'].astype(str).str.split('.').str.get(1)
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df['ball_no'] = df['ball_no'].astype(int)
```

```python
df['total_delivers'] = (df['over']*6) + df['ball_no']
```

```python
df['balls_left'] = 120 - df['total_delivers']
```

```python
df['balls_left'].mask(df['balls_left'] < 0, 0, inplace=True)
```

```python
df
```

```python
df['player_dismissed'] = df['player_dismissed'].mask(df['player_dismissed'] != '0', 1).astype(int)
```

```python
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

### Run Rate

```python
df['crr'] = df['current_score']*6/df['total_delivers']
```

```python
df.head(10)
```

### Last 5 overs

```python
df['last_five'] = df.groupby('match_id')['runs'].rolling(window=30).sum().to_list()
```

```python
df.head(40)
```

```python
total_run_table = df.groupby('match_id')['runs'].sum().reset_index()
```

```python
df = df.merge(total_run_table, on='match_id')
```

```python
df
```

```python
df.columns
```

```python
final_df = df[['batting_team','bowling_team','current_score' , 'balls_left', 'wickets_left', 'crr', 'last_five', 'runs_y']]
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
X = final_df.drop(columns=['runs_y'])
y = final_df['runs_y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = 1)
```

```python
y_train = y_train.squeeze()
y_test = y_test.squeeze()
```

```python
trf = ColumnTransformer([('trf',OneHotEncoder(sparse=False), ['batting_team','bowling_team'])], remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', StandardScaler()),
    ('step3', XGBRegressor(n_estimators = 1000, learning_rate = 0.2, max_depth = 12, random_state= 1))
])
```

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
```

```python
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe, open('pipe.pkl','wb'))
```
